## Creating our custom data-Loader

```python
def idx2label():
    id2label={0:"mismatch",1:"match"}
    print(id2label)
    return id2label


class MyCustomDataset(torch.utils.data.Dataset):
    def __init__(self,root,transform=None):
        self.root_dir=root
        self.transform=transform
        
        self.similar_pairs = os.path.join(root,"match_pairs")
        self.similar_pairs_folders=os.listdir(self.similar_pairs)
        
        self.different_pairs = os.path.join(root,"mismatch_pairs")
        self.different_pairs_folders=os.listdir(self.different_pairs)
        
    def __len__(self):
        return len(self.similar_pairs_folders)+len(self.different_pairs_folders)
    
    def __getitem__(self, idx):
           if idx<len(self.similar_pairs_folders):
               pairs_path=os.path.join(self.similar_pairs,f"sim_pair{idx}")
               img1=os.path.join(pairs_path,os.listdir(pairs_path)[0])
               img2=os.path.join(pairs_path,os.listdir(pairs_path)[1])
               img1=PIL.Image.open(img1)
               img2=PIL.Image.open(img2)
               label=1
           else:
               pairs_path=os.path.join(self.different_pairs,f"mismatch_pair{idx-len(self.similar_pairs_folders)}")
               img1=os.path.join(pairs_path,os.listdir(pairs_path)[0])
               img2=os.path.join(pairs_path,os.listdir(pairs_path)[1])
               img1=PIL.Image.open(img1)
               img2=PIL.Image.open(img2)
               label=0
           if self.transform:
            img1=self.transform(img1)
            img2=self.transform(img2)
            
           return img1,img2,label


```

```python
transform=Compose([ToTensor(),
                   Normalize((.5,.5,.5),(.5,.5,.5),inplace=True),
                   Resize((224,224))])
dataset=MyCustomDataset(root="dataset/",transform=transform)
```

```python
# dividing into training and val test (stratified split)
from  sklearn.model_selection import train_test_split
labels=[1]*len(dataset.similar_pairs_folders)+[0]*len(dataset.different_pairs_folders)

train_indices,test_indices = train_test_split(
                          range(len(labels)),
                          test_size=0.3,
                          stratify=labels)
```

```python
train_dataset=torch.utils.data.Subset(dataset,indices=train_indices)
test_dataset=torch.utils.data.Subset(dataset,indices=test_indices)
```

```python
# creating data loaders
batch=32
train_loader=DataLoader(train_dataset,shuffle=True,batch_size=batch,num_workers=0)
test_loader=DataLoader(test_dataset,shuffle=True,batch_size=batch,num_workers=0)
```

```python
# visualize the classes distributions for both training and val dataset
def visualize_class_distribution(dataset):
    ones=0
    zeros=0
    for example in dataset:
        if example[2] == 1:
            ones += 1
        else:
            zeros += 1
    
    plt.bar(x=["similar","different"],height=[ones,zeros],align="center")
    plt.tight_layout()
    plt.show()
```

## Implementing the model  

```python
from transformers import ViTModel

class Siamese_NN(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder=ViTModel.from_pretrained("google/vit-base-patch16-224")
        self.linear_layer=torch.nn.Linear(2*768,128)
        self.relu=torch.nn.ReLU()
        self.classification_head=torch.nn.Linear(128,1)
        self.sigmoid=torch.nn.Sigmoid()
        
    def forward(self,img1,img2):
        img1,img2=img1.to(device),img2.to(device),
        x1=self.encoder(img1).last_hidden_state[:,0,:]
        x2=self.encoder(img2).last_hidden_state[:,0,:]
        concatenated = torch.cat((x1, x2), dim=1)
        outputs=self.relu(self.linear_layer(concatenated))
        outputs=self.sigmoid(self.classification_head(outputs))
    
        return outputs
        
        
    
```

```python
# setting the device to cuda 
device=torch.device("cuda:0") if torch.cuda.is_available() else torch.device("cpu")
device
```

```python
model=Siamese_NN().to(device)
```

```python
for name, param in model.encoder.named_parameters():
    # Set requires_grad=False for all parameters of the encoder except model.encoder.pooler.dense
    print(name)
    if name not in ["pooler.dense.weight","pooler.dense.bias"]:
        param.requires_grad = False
```

```python
def print_model_params(model):
    num_params = 0
    layers = [layer for layer in model.parameters() if layer.requires_grad]
    for i, layer in enumerate(layers, start=1):
        print(f"Trainable layer {i}: {layer.shape}")
        num_params += layer.numel()
    print(f"Number of trainable parameters: {num_params}")
    return num_params
```

```python
print_model_params(model)
```

```python
# define optimizer and loss function 
opt=torch.optim.AdamW(params=model.parameters(),lr=1e-3)
criterion=torch.nn.BCELoss()
```

```python
def train_one_epoch(model,criterion,optimizer,data_loader):
    total_loss = 0
    size=0
    total_correct = 0
    total_size = len(data_loader.dataset)
    total_labels=[]
    total_predictions=[]
    model.train()
    avg_score=0
    history={
        "metrics":{
            "accuracy":[],
            "f1_score":[]
                  },
        "loss":[]   
            }  
    for batch_number,batch in enumerate(data_loader):
        print(f"batch number {batch_number}/{len(data_loader)}")
        x1,x2,labels=batch[0].to(device),batch[1].to(device),torch.unsqueeze(batch[2].to(device),dim=-1).float()
        optimizer.zero_grad()
        # forward pass 
        probabilities = model(x1,x2)
        # compute the loss
        loss=criterion(probabilities,labels)
        #computes gradient
        loss.backward()
        optimizer.step()
        total_loss+=loss.item()
    
        # compute metrics
        predicted_labels  = (probabilities > 0.8).int()
        total_correct += (predicted_labels == labels).int().sum()
        size+=len(labels)
        predicted_labels_cpu=copy.copy(predicted_labels).cpu().numpy()
        labels_cpu=copy.copy(labels).cpu().numpy()

        total_labels.extend(labels_cpu)
        total_predictions.extend(predicted_labels_cpu)
        
        
        if batch_number % 32== 0:
            accuracy=(total_correct/size)
            score=f1_score(total_labels,total_predictions)
            print(f"loss {loss.item()} accuracy { accuracy * 100} f1-score {score} ")
            history["metrics"]["accuracy"].append(accuracy.float().cpu())
            history["metrics"]["f1_score"].append(score)
            history["loss"].append(loss)
    print(f"Total accuracy {(total_correct/size) *100}")
    print(f"Average Loss: {total_loss / len(data_loader)}")
    print(f"Average f1_score {f1_score(total_labels,total_predictions)}")
    #saving accuracy and loss for visualization
    return history   
```

```python
def test(model,criterion,data_loader):
    total_loss=0
    model.eval()
    total_correct=0
    size=0
    total_labels=[]
    total_predictions=[]
    history={
        "metrics":{
            "accuracy":[],
            "f1_score":[]
        },
        "loss":[]
    }
    with torch.no_grad():
        print("Beginning Testing ")
        for batch_number,batch in enumerate(data_loader):
            x1,x2,labels=batch[0].to(device),batch[1].to(device),torch.unsqueeze(batch[2].to(device),dim=-1).float()
            outputs = model(x1,x2)
            loss=criterion(outputs,labels).item()
            predicted_labels= (outputs>0.8).int()
            correct=(predicted_labels==labels).int().sum()
            total_correct+=correct
            total_loss+=loss
            size+=len(labels)
            
            predicted_labels_cpu=copy.copy(predicted_labels)
            predicted_labels_cpu=predicted_labels.cpu().numpy()
            labels_cpu=copy.copy(labels)
            labels_cpu=labels_cpu.cpu().numpy()
            
            total_labels.extend(labels_cpu)
            total_predictions.extend(predicted_labels_cpu)
            
            accuracy=total_correct/size
            if batch_number % 32== 0:
                score=f1_score(total_predictions,total_labels)
                history["metrics"]["accuracy"].append(accuracy.float().cpu())
                history["metrics"]["f1_score"].append(score)
                history["loss"].append(loss)
                print(f"loss {loss} accuracy { accuracy * 100} f1-score {score} ")
        avg_loss=total_loss/len(data_loader)
        avg_score=f1_score(total_predictions,total_labels)
        print(f"Finished testing! Model Evaluation: Accuracy at {accuracy} ---- Loss {avg_loss} --- f1_score {avg_score}")
    return history
```

```python
def save_logs(data_log,history):
         for metrics_type,metrics in data_log["metrics"].items():
             history["metrics"][metrics_type].extend(metrics)
         history["loss"].extend(data_log["loss"])
```

```python
history={"training_log":{
                "metrics":{
                    "accuracy":[],
                    "f1_score":[]
                           },
                "loss":[]
                },
         "test_log":{
                "metrics":{
                    "accuracy":[],
                    "f1_score":[]
                          },
                 "loss":[]
                }
        }
for epoch in range(10):
    print(f"epoch {epoch} {'-'*100}" )
    training_log = train_one_epoch(model,criterion,opt,train_loader)
    test_log = test(model,criterion,test_loader)
    
    save_logs(training_log,history["training_log"])
    save_logs(test_log,history["test_log"])
        
```

# VGG

```python
# implementing the SNN with VGG encoder
class SNN_with_vgg(torch.nn.Module):
    def __init__(self):
        super().__init__()
        vgg=torchvision.models.vgg16().features
        self.vgg=torch.nn.Sequential(*list(vgg.children()))
        self.encoder=torch.nn.Sequential(
                                torch.nn.Flatten(),
                                torch.nn.Linear(25088,4096),
                                torch.nn.ReLU(inplace=True),
                                torch.nn.Dropout(0.5),
                                torch.nn.Linear(in_features=4096,out_features=768)
                                )
        self.distance=torch.nn.PairwiseDistance()
        

    def forward(self,img1,img2):
        img1,img2=img1.to(device),img2.to(device),
        x1=self.encoder(self.vgg(img1))
        x2=self.encoder(self.vgg(img2))
        d=self.distance(x1,x2)
        d=torch.unsqueeze(d,dim=1)
        return d

```

```python
class ContrastiveLoss(torch.nn.Module):
    """
    Contrastive loss function.
    """

    def __init__(self, margin=2.0):
        super(ContrastiveLoss, self).__init__()
        self.margin = margin

    def forward(self, dist, label):

        loss = torch.mean(1/2*(label) * torch.pow(dist, 2) +
                          1/2*(1-label) * torch.pow(torch.clamp(self.margin - dist, min=0.0), 2))


        return loss
```

```python
criterion_2=ContrastiveLoss()
optimizer_2=torch.optim.AdamW(vgg_SNN.parameters(),lr=1e-3)
```

```python
def train_vgg_one_epoch(model,criterion,optimizer,data_loader):
    history={
        "accuracy":[],
        "loss":[]
            }
    size=0
    total_loss=0
    total_correct=0
    for batch_number,batch in enumerate(data_loader,start=1):
        print(f"batch number {batch_number}/{len(data_loader)}")
        x1,x2,labels=batch[0].to(device),batch[1].to(device),torch.unsqueeze(batch[2].to(device),dim=-1).float()
    
        optimizer.zero_grad()
        # forward pass 
        d = model(x1,x2)
        # compute the loss
        loss=criterion(d,labels)
        #computes gradient
        loss.backward()
        optimizer.step()
        total_loss+=loss.item()

        # compute metrics by applying a threshold
        predicted_labels= (d <= 0.5).int()
        
       
       
        
        total_correct += (predicted_labels == labels).int().sum()
        size+=len(labels)
        


        if batch_number % 32== 0:
            accuracy=(total_correct/size)
            print(f"loss {total_loss/batch_number} accuracy { accuracy * 100} ")
            history["accuracy"].append(accuracy.float().cpu())
            history["loss"].append(loss)
            
            
    print(f"Total accuracy {(total_correct/size) *100}")
    print(f"Average Loss: {total_loss / len(data_loader)}")
    #saving accuracy and loss for visualization
    return history   
```

```python
def test_vgg(model,criterion,data_loader):
    total_loss=0
    total_correct=0
    size=0
    total_labels=[]
    total_predictions=[]
    history={
            "accuracy":[],
            "loss":[]
            }

    model.eval()
    with torch.no_grad():
        print("Beginning Testing ")
        for batch_number,batch in enumerate(data_loader,start=1):
            x1,x2,labels=batch[0].to(device),batch[1].to(device),torch.unsqueeze(batch[2].to(device),dim=-1).float()
            d = model(x1,x2)
            loss=criterion(d,labels).item()
            predicted_labels= (d <= 0.5).int()
            correct=(predicted_labels==labels).int().sum()
            total_correct+=correct
            total_loss+=loss
            size+=len(labels)

            accuracy=total_correct/size
            if batch_number % 32== 0:
                
                history["accuracy"].append(accuracy.float().cpu())
                history["loss"].append( total_loss/batch_number )
                print(f"loss {loss} accuracy { accuracy * 100}  ")
        avg_loss=total_loss/len(data_loader)
        print(f"Finished testing! Model Evaluation: Accuracy at {accuracy} ---- Loss {avg_loss}")
    return history
```

```python
history={
    "training_log":{
        "accuracy":[],
        "loss":[]
                    },
    "test_log":{
        "accuracy":[],
        "loss":[]
               }
    }

```

```python
for epoch in range(10):
    print(f"epoch {epoch} {'-'*100}" )
    training_log = train_vgg_one_epoch(vgg_SNN,criterion_2,optimizer_2,train_loader)
    test_log = test_vgg(vgg_SNN,criterion_2,test_loader)

    history["training_log"]["accuracy"].extend(training_log["accuracy"])
    history["training_log"]["loss"].extend(training_log["loss"])
    history["test_log"]["accuracy"].extend(test_log["accuracy"])
    history["test_log"]["loss"].extend(test_log["loss"])
```
